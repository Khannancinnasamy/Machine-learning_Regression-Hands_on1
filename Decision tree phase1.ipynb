```python
# Pandas is a library for Data manipulation
import pandas as pd
```

```python
df = pd.read_csv(r"50_Startups.csv")
```

```python
df
```

```python
## Convert categorical columns to numerical using one hot encoding as the state is nominal data
df = pd.get_dummies(df, drop_first = True).astype(int)
```

```python
df
```

```python
df.columns
```

```python
# input data split
independent = df[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']]
```

```python
# Output data split
dependent= df[['Profit']]
```

```python
# Train and test data split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(independent, dependent, test_size = 0.30, random_state = 0)
```

```python
X_train
```

```python
# Scale the data to fix the particular range
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Building the model for Decisiontree Regression
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor()
regressor = regressor.fit(X_train, y_train)
```

```python
regressor
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score = r2_score(y_test, y_pred)
```

```python
r_score
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor(criterion='friedman_mse', splitter='random', random_state = 0)
regressor=regressor.fit(X_train,y_train)
```
